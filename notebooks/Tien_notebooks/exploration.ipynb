```python
# List of raw variables based on your selected fields
raw_fields = [
    'w1_ppage', 'w1_q9',              # Ages
    'w1_ppeduc', 'w1_q10',            # Education
    'w1_ppincimp',                    # Household Income
    'w1_partyid7', 'w1_q12',          # Politics
    'w1_ppp20072',                    # Religion
    'w1_ppethm', 'w1_q6b',            # Race/Ethnicity
    'w1_ppmsacat', 'w1_ppreg9',       # Geography
    'w1_q24_met_online',              # Meeting Context
    'w1_weight_combo'                 # Weights (Crucial!)
]

# Create a clean working copy
data = df[raw_fields].copy()
```

```python
# 1. Age Gap (Absolute value)
data['age_gap'] = np.abs(pd.to_numeric(data['w1_ppage'], errors='coerce') - 
                         pd.to_numeric(data['w1_q9'], errors='coerce'))

# 2. Education Match (Binary)
# Using .astype(str) ensures we compare the labels ('College', 'High School') correctly
data['edu_match'] = (data['w1_ppeduc'].astype(str) == data['w1_q10'].astype(str)).astype(int)

# 3. Political Match (Binary)
data['pol_match'] = (data['w1_partyid7'].astype(str) == data['w1_q12'].astype(str)).astype(int)

# 4. Race Match (Binary)
data['race_match'] = (data['w1_ppethm'].astype(str) == data['w1_q6b'].astype(str)).astype(int)

# 5. Religious Intensity (Ordinal)
# We convert categories (Never, Once a week, etc.) into a 0-6 scale
data['relig_intensity'] = data['w1_ppp20072'].cat.codes

# 6. Meeting Context (Binary)
data['is_online'] = data['w1_q24_met_online'].astype(str).str.contains('met online', case=False).astype(int)

# 7. Income level (Ordinal)
data['income_level'] = data['w1_ppincimp'].cat.codes
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Selecting the engineered features for the model
clustering_vars = ['age_gap', 'edu_match', 'pol_match', 'race_match', 
                   'relig_intensity', 'is_online', 'income_level']

X = data[clustering_vars]

# Handle missing values (using median is safest for these types of variables)
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)

# Standardize features (mean=0, variance=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
from sklearn.cluster import KMeans

# Using 4 clusters to find broad "Relationship Tribes"
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
data['cluster'] = kmeans.fit_predict(X_scaled)

# Calculate cluster centers to understand the "Story"
# We multiply back by the weights for more "honest" averages
def weighted_mean(x, w):
    return np.average(x, weights=w)

# Group by cluster and look at the profiles
profile_summary = data.groupby('cluster').apply(
    lambda x: pd.Series({
        'Avg Age Gap': np.mean(x['age_gap']),
        '% Political Match': np.mean(x['pol_match']) * 100,
        '% Met Online': np.mean(x['is_online']) * 100,
        'Relig. Intensity (0-6)': np.mean(x['relig_intensity']),
        '% Race Match': np.mean(x['race_match']) * 100
    })
)
print(profile_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))
sns.heatmap(profile_summary.T, annot=True, cmap="YlGnBu", fmt=".1f")
plt.title('Relationship Profile Archetypes (Part 1)')
plt.show()
```
